```
import os 
import pandas as pd
import numpy as np 
import seaborn as sns
%matplotlib inline


from patsy import dmatrices
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import f1_score, classification_report, accuracy_score, mean_squared_error
```

```
# reading in Dataset
paths = 'http://bigdata.cs.byu.edu/Lectures/EEGEyeState.csv'
df = pd.read_table(paths,sep=',',encoding='latin1')

# split dependent and independent variables
y = df['eyeDetection']
X = df.iloc[:,:-1]

# create test and train.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.3,random_state=40)
```

```
# fit KNN

clf = KNeighborsClassifier()
gridsearch = GridSearchCV(clf, {"n_neighbors": [1, 3, 5, 7], "weights": ['uniform', 'distance'], 
                                'p': [1, 2, 3]}, scoring='f1')
gridsearch.fit(X_train, y_train)
`
```

```
#best parameters
print("Best Params: {}".format(gridsearch.best_params_))
y_pred_train = gridsearch.predict(X_train)
#classification report
print("Test Classification Report:")
y_pred_test = gridsearch.predict(X_test)
print(classification_report(y_test, y_pred_test))


# Confusion Matrix
#print("Confusion Matrix")
conf_mat = pd.crosstab(y_test,y_pred_test,margins=True)
#print(conf_mat)


#sensitivity 
print('Our Sensitivity value is:', conf_mat.iloc[1,1]/conf_mat.iloc[1,2])
#specificity
print('Our Specificity is:',conf_mat.iloc[0,0]/conf_mat.iloc[0,2])
#positive predictive value 
print('Our PPV is:',conf_mat.iloc[1,1]/conf_mat.iloc[2,1])
#negative predictive value 
print('Our NPV is:',conf_mat.iloc[0,0]/conf_mat.iloc[2,0])
#accuracy
print("Test accuracy: {}".format(accuracy_score(y_test, y_pred_test)))
```
