```
import os 
import pandas as pd
import numpy as np 
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from patsy import dmatrices
import statsmodels.api as sm
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix 
```

## Reading in Dataset

```
# reading in Dataset
paths = 'http://bigdata.cs.byu.edu/Lectures/EEGEyeState.csv'
df = pd.read_table(paths,sep=',',encoding='latin1')

# split dependent and independent variables
X = list(df)[:-1]
y = ['eyeDetection']

# split into dataframe
y, X = dmatrices('eyeDetection~AF3+F7+F3+FC5+T7+P7+O1+O2+P8+T8+FC6+F4+F8+AF4',df,return_type='dataframe') 

# create test and train.
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.1,random_state=40)
```

# Random Forest

```
rf = RandomForestClassifier()
param = {'n_estimators': [100, 150, 250],
        'max_depth': [10, 30, 50]}

gs = GridSearchCV(rf, param, cv=5, n_jobs=-1)
gs_fit = gs.fit(X_train, y_train)
pd.DataFrame(gs_fit.cv_results_).sort_values('mean_test_score', ascending=False)
```

```
testnums = [0, 1, 2]

# select param_n_estimators as 100
# vary maximum depth by 10, 30, 50.
params = [
    [100, 50],
    [100, 30],
    [100, 10]
]

for test in testnums:
  random_forest = RandomForestClassifier(n_estimators=params[test][0], max_depth=params[test][1],
                             random_state=7, max_features = .33 ,oob_score=True)
  random_forest.fit(X_train, np.ravel(y_train))

  print("Accuracy:")
  print(random_forest.score(X_test, y_test))

  print("Importances:")
  print(sorted(zip(random_forest.feature_importances_, X_test.columns), reverse=True))
  
  print("Compliment OOB Error")
  print(random_forest.oob_score_)
  
  print("Confusion Matrix")
   
  cmat = confusion_matrix(y_test, random_forest.predict(X_test))
  mats = pd.DataFrame(cmat)
  print(mats)
  
  print("Sensitivity")
  print(mats.iloc[1,1]/( mats.iloc[1,1]+ mats.iloc[0,1]))
  print("Specificity")
  print(mats.iloc[0,0]/( mats.iloc[0,0]+ mats.iloc[0,1]))
```

```
ran_forest = RandomForestClassifier(n_estimators=100, max_depth=50,
                             random_state=7, max_features = .33 ,oob_score=True)
ran_forest.fit(X_train, np.ravel(y_train))


print("Accuracy:")
print(ran_forest.score(X_test, y_test))

print("Importances:")
print(sorted(zip(ran_forest.feature_importances_, X_test.columns), reverse=True))
  
print("Compliment OOB Error")
print(ran_forest.oob_score_)
  
print("Confusion Matrix")
   
cmat = confusion_matrix(y_test, ran_forest.predict(X_test))
mats = pd.DataFrame(cmat)
print(mats)
  
print("Sensitivity")
print(mats.iloc[1,1]/( mats.iloc[1,1]+ mats.iloc[0,1]))
print("Specificity")
print(mats.iloc[0,0]/( mats.iloc[0,0]+ mats.iloc[0,1]))
```

This result suggests 14% and greater of the predicted values in the "out of sample" data (unused in in the creation of the tree), is accurate. Furthermore, since our values of .oob_score_ fall way beyond 50% (by 46) our ensemble outperforms just flipping a coin by more than 46%, 

Furthermore, using an alternative Accuracy scoring method, our best gridsearch cross validation metrics indicate high accuracy with a maximum depth of 50 (92%). The result indicates a high percentage of correct prediction out of total number of predicted values. This fold had the lowest estimate of overfitting compared to all other cross validated random forest models.
