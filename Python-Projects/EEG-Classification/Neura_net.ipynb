```
import pandas as pd  
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
```

```
paths = 'http://bigdata.cs.byu.edu/Lectures/EEGEyeState.csv'
df = pd.read_table(paths,sep=',',encoding='latin1')
```

```
y = df['eyeDetection']
X = df.iloc[:,:-1]
```

```
print(len(y))
print(len(X.O1))
```

```
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.3,random_state=40)
```

```
scaler = StandardScaler()
```

```
scaler.fit(X_train)
```

```
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```
mlp = MLPClassifier(activation='relu',hidden_layer_sizes=(12,12,12),max_iter=500)
```

```
mlp.fit(X_train,y_train)
```

```
nn_preds = mlp.predict(X_test)
```

```
conf_mat = pd.crosstab(y_test,nn_preds,margins=True)
conf_mat
```

```
#sensitivity 
print('Our Sensitivity value is:', conf_mat.iloc[1,1]/conf_mat.iloc[1,2])
#specificity
print('Our Specificity is:',conf_mat.iloc[0,0]/conf_mat.iloc[0,2])
#positive predictive value 
print('Our PPV is:',conf_mat.iloc[1,1]/conf_mat.iloc[2,1])
#negative predictive value 
print('Our NPV is:',conf_mat.iloc[0,0]/conf_mat.iloc[2,0])
#accuracy
print('Our accuracy score is:',accuracy_score(y_test,nn_preds))
```

```

roc_auc_score(y_test, nn_preds)
```

# Summary

Our neural network results indicate an accuracy over 55%. This suggests fewer than 45% accuracy when predicting values out of our trained data. 

From the confusion matrix we observe specificity near 76%, indicating the ability to predict eye state when it is not open. Sensitivity, on the other hand, was found to be near 30% suggesting the modelâ€™s capacity to predict whether the eye was open when it truly was. As the aim is to maximize sensitivity, 30% may not be the optimum model for this task.

Another metric that is used to measure the performance of this model is the area under the curve: 56%. Though this is slightly better than a coin flip, this particular neural network model does not perform as well as other models for this particular analysis. This may be due to the simple composition of the network with 3 layers and 12 hidden nodes. For this model to determine utility, a more complex model (ie. RNN) may need to be considered.
